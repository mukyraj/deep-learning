```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load and Preprocess Your Dataset
df = pd.read_csv('seattle-weather.csv')  # Replace 'seattle-weather.csv' with the actual file path

# Drop rows with null values
df.dropna(inplace=True)

# Encode labels
label_encoder = LabelEncoder()
df['weather_condition_encoded'] = label_encoder.fit_transform(df['weather'])

# Extract features and labels
X = df[['precipitation', 'temp_max', 'temp_min', 'wind']].values
y = df['weather_condition_encoded'].values

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform SMOTE for balancing classes
smote = SMOTE()
X_balanced, y_balanced = smote.fit_resample(X_scaled, y)

# Split the balanced data into training and testing sets
X_train_balanced, X_test_balanced, y_train_balanced, y_test_balanced = train_test_split(X_balanced, y_balanced, test_size=0.2, random_state=42)

# Define the model with adjusted architecture
model_balanced = Sequential([
    LSTM(units=64, input_shape=(X_train_balanced.shape[1], 1), return_sequences=True),
    Dropout(0.3),
    LSTM(units=32),
    Dense(units=len(np.unique(y)), activation='softmax')  # Output layer with softmax activation
])

# Compile the model with class weights
class_weights_balanced = dict(enumerate((len(y_train_balanced) / (len(np.unique(y_train_balanced)) * np.bincount(y_train_balanced))).tolist()))
model_balanced.compile(optimizer=Adam(learning_rate=0.01), loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model with class weights
history_balanced = model_balanced.fit(X_train_balanced[..., np.newaxis], y_train_balanced, epochs=50, batch_size=64, validation_split=0.2, class_weight=class_weights_balanced)

# Evaluate the model
loss_balanced, accuracy_balanced = model_balanced.evaluate(X_test_balanced[..., np.newaxis], y_test_balanced)
print("Test Loss (Balanced):", loss_balanced)
print("Test Accuracy (Balanced):", accuracy_balanced)

# Generate Classification Report for balanced data
y_pred_balanced = model_balanced.predict(X_test_balanced[..., np.newaxis])
y_pred_classes_balanced = np.argmax(y_pred_balanced, axis=1)
class_names = label_encoder.inverse_transform(np.unique(y))  # Convert encoded classes back to weather conditions

# Print classification report for balanced data
print(classification_report(y_test_balanced, y_pred_classes_balanced, target_names=class_names))

# Plot confusion matrix
cm = confusion_matrix(y_test_balanced, y_pred_classes_balanced)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix (Balanced Data)')
plt.show()

# Plot actual vs predicted for balanced data using a bar graph
plt.figure(figsize=(10, 6))
index = np.arange(len(y_test_balanced))
bar_width = 0.35
plt.bar(index, y_test_balanced, bar_width, label='Actual')
plt.bar(index + bar_width, y_pred_classes_balanced, bar_width, label='Predicted')
plt.xlabel('Sample Index')
plt.ylabel('Weather Condition')
plt.title('Actual vs Predicted Weather Condition (Balanced Data)')
plt.xticks(index + bar_width / 2, index)
plt.legend()
plt.show()
```
